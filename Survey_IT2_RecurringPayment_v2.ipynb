```
# Library
library(ggplot2)
library(tidyverse)
#install.packages("readxl")
library(readxl)
```

```
# reading in input File
input_file = '/content/Without - Payment Pref - Recurring.xlsx'
df_master = read_xlsx(input_file, sheet=1)
```

```
# printing DF
print(as_tibble(df_master))
```

```
names(df_master)
```

removing NA

```
df_master %>% replace_na(list('12-Months' = 0,  '3-Months' = 0, '6-Months' = 0, '1-Months' = 0)) -> df_1
```

```
str(df_1)
```

```
write.csv(df_1,"recurring.csv")
```

```
df_1 %>%
  pivot_longer(!Amount, names_to = "method", values_to = "Count") -> df_2
```

```
write.csv(df_2,"recurring2.csv")
```

```
str(df_2)
```

converting character to factor type

```
#one approach it to index with the $ sign and the as.factor function
df_2$method_f <- as.factor(df_2$method)
```

```
p1 <- ggplot(df_2, aes(x = Amount, y = Count))
# Scatter plot 
p1 <- p1 + geom_point()
p1
```

```
p1 <- p1 + theme_minimal()
p1
```

```
# bars are too little to add data labels
p1 <- p1 + labs(x = "RECURRING PAYMENT (PKR)", y = "COUNT")
p1
```

```
# Split in vertical direction
p1 + facet_grid(method_f ~ .)
```

```
# removing outliers of any value above 25000
df_3 <- df_2 %>% filter(Amount <= 25000)
```

```
p10 <- ggplot(df_3, aes(x = Amount, y = Count))
# Scatter plot 
p10 <- p10 + geom_point()
p10 <- p10 + theme_minimal()
p10
```

```
p25 <- ggplot(df_3, aes(x = Amount, y = Count))
# Scatter plot 
p25 <- p25 + geom_point(aes(color=method_f),pch = 21, position = position_jitterdodge())
p25 <- p25 + theme_minimal()
#p25 <- p25 + theme(legend.position = "none")
p25 <- p25 + theme(legend.position = c(0.6, 0.8),
          legend.direction = "vertical")

# Edit legend title and labels
p25 <- p25 + scale_color_discrete(name = "Recurrence")
p25
```

```
# remove legend for alpha
p25 <- p25 + guides(alpha = FALSE)
p25
```

```
p25 <- p25 + labs(x = "RECURRING PAYMENT (PKR)", y = "COUNT")
p25
```

```
# Split in vertical direction, MARGINS gets all values
p25 <- p25 + theme(legend.position = "none")
p30 <- p25 + facet_grid(method_f ~ ., margins=TRUE)
```

```
p30
```
